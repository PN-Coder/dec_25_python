```python
import pandas as pd
import random # importing libraries required to run some in-buit methods and functions

master_file = pd.read_csv('/content/drive/MyDrive/Regions_Capital_File.csv') # bringing file into the program
print(master_file) # checking if all the values are successfully extracted from csv
```

```python
regions_list = master_file['Country'].to_list() # extracting column into list so can use them

capitals_list = master_file['Capital'].to_list() # extracting column into list so can use them
```

```python
marks = 0                      # to keep track of score

question_count = 1                            # question number, ideally we can start index from zero, but in BAU, we say question no. 1, that is why starting from 1
question_country_list = []                    # this list will store the countries on which question have been asked, to avoid repetitive questions
while question_count<6:
  question_country = random.choice(regions_list)                  # a member will be selected randomly from the regions_list


  if question_country not in question_country_list:               # This if-else statement will check if the country we have randomly selected is already in the previous questions, if it is, then it will select another value
    question_country_list.append(question_country)
    question_count = question_count + 1
  else:
    continue



  ideal_answer = master_file.loc[master_file['Country']==question_country]['Capital'].iloc[0]                # Code on this line will extract ideal answer for the question country we are about to ask
  temp_ans_list = []                             # This list will contain all the values in the capitals_list except the correct one, we will use this to prepare three extra values to fill up our multiple choice
  for i in capitals_list:
    if i == ideal_answer:
      continue
    else:
      temp_ans_list.append(i)


  print('What is the caital of',question_country)                   # asking the question

  index = random.randint(0,3)                     # this is the random index value among 1 to 4 (A to D) options where we rotate the correct option
  answer_options_list = []                     # list that contains all the options correct and incorrect
  i = 0
  while i<4:
    other_option = random.choice(temp_ans_list)           # selects options other than correct ones
    if i != index and other_option not in answer_options_list:        # to avoid repetition of incorrect options
      answer_options_list.append(other_option)
      i = i+1
    elif i == index:
      answer_options_list.append(ideal_answer)           # fill out the correct option in the options list
      i = i+1
    else:
      continue


  options_list_upper = ['A','B','C','D']              # if user enters the response in upper case
  options_list_lower = ['a','b','c','d']              # if user enters the response in lower case
  for i in range(0,4):
    print(options_list_upper[i],answer_options_list[i])
  user_input = input('Enter among the options, the correct one\n')

  if user_input in options_list_upper:
    if (options_list_upper.index(user_input) == answer_options_list.index(ideal_answer)):
      print('Well Done, you selected correct option')
      marks = marks+1
    else:
      print('Good Try, but wrong one')
  elif user_input in options_list_lower:
    if options_list_lower.index(user_input) == answer_options_list.index(ideal_answer):
      print('Well Done, you selected correct option')
      marks = marks + 1
    else:
      print('Good Try, but wrong one')

print('You have scored',marks,'out of 5')                   # printing the final result
```
